```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X = df2.drop('Survived', axis =1)
y = df2['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
model = LogisticRegression(random_state = 42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])

disp.plot(cmap=plt.cm.Blues) # You can change the colormap
plt.title('Confusion Matrix')
plt.show()
```

```python
import seaborn as sns

y_pred_proba = model.predict_proba(X_test)[:, 1]
plt.figure(figsize=(10, 6))
sns.histplot(y_pred_proba[y_test == 0], color='red', label='Actual Class 0 (Negative)', kde=True, stat="density", linewidth=0)
sns.histplot(y_pred_proba[y_test == 1], color='blue', label='Actual Class 1 (Positive)', kde=True, stat="density", linewidth=0)
plt.axvline(x=0.5, color='green', linestyle='--', label='Decision Boundary (0.5)') # Default threshold for logistic regression
plt.title('Distribution of Predicted Probabilities by Actual Class')
plt.xlabel('Predicted Probability of Positive Class')
plt.ylabel('Density')
plt.legend()
plt.show()
```
