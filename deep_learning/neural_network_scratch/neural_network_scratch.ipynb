# Activation Functions

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def softmax(z):
    expZ = np.exp(z)
    return expZ/(np.sum(expZ, 0))

def relu(Z):
    A = np.maximum(0, Z)
    return A

def tanh(x):
    return np.tanh(x)

def derivative_relu(x):
    return np.array(x>0, dtype = 'float')

def derivative_tanh(x):
    return (1 - np.power(x, 2))
```
